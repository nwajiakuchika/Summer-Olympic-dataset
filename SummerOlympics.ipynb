```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Reading of file

```python
df = pd.read_csv("Athletes_summer_games.csv")
df
```

# Droping the first column

```python
df = df.drop(columns=['Unnamed: 0'])
df
```

# Information of dataset

```python
df.info()
```

# Handling mising medal values

```python
df['Medal'] = df['Medal'].fillna("No Medal")
df
```

# Numbers of athletes per year

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Year', data=df)
plt.xticks(rotation=90)
plt.title("Number of Athletes per Olympic Year")
plt.show()
```

# Medal distribution

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Medal', data=df)
plt.title("Medal Distribution")
plt.show()
```

# Gender participant over the years

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Year', hue='Sex', data=df)
plt.xticks(rotation=90)
plt.title("Male vs Female Participation Over the Years")
plt.show()
```
